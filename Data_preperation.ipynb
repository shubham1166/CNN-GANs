The images are 8 bit images of shape(100, 100, 3). Lets make x_train and x_val

```
#Let us first make X
dim = (256, 256)
X = np.zeros([len(list_of_images) , 3 , 256 , 256])
for i in tqdm(range(len(list_of_images))):
  img = cv2.imread(path2data+list_of_images[i],1)
  if img.shape[0] != 256 or img.shape[1] != 256:
    img = cv2.resize(img, dim)
  im_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB) # As cv2 lod in BGR format and matplotlib takes input as RGB
  im_rgb = im_rgb/255.
  #Converting to channel first
  im_rgb = np.transpose(im_rgb, axes=(2,0,1))
  X[i] = im_rgb

  
```

```
x_train = X[:788]
x_val = X[788:]
```
